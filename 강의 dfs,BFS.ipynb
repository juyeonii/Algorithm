# DFS(깊이 우선 탐색)
재귀,스택 사용\
낮은 번호가 우선\
최상단 노드의 방문하지 않은 노드를 탐색하면서 깊이 들어감


```python
def dfs(graph,v,visited):
    visited[v]=True
    print(v,end=' ')
    for i in graph[v]:
        if not visited[i]:
            dfs(graph,i,visited)
```

# BFS(너비 우선 탐색)
가까운 노드부터 우선적으로 탐색\
큐\
방문하지 않은 노드를 모두 큐에 삽입하고 방문\
최단경로문제

```python
from collections import deque
def bfs(graph,start,visited):
    queue=deque([start])
    visited[start]=True
    while queue:
        v=queue.popleft()
        print(v,end=' ')
        for i in graph[v]:
            if not visited[i]:
                visited[i]=True
```

<문제> 음료수 얼려먹기

#### dfs
1.상하좌우 주변을 살펴본 뒤 지점 중 값이 0이면서 아직 방문하지 않으면 방문\
2.방문한 지점에서 상하좌우 살펴보고 방문을 진행하는 과정을 반복->모두 방문\
3.1-2과정을 반복하며,방문하지 않은 지점의 수를 카운트

```python
N,M=map(int,input().split())
graph=[]
for _ in range(N):
    graph.append(list(map(int,input().split())))
```

```python
def dfs(x,y):
    if x<0 or y<0 or x>=N or y>=M:
        return False #종료
    if graph[x][y]==0:
        graph[x][y]=1
        dfs(x-1,y)
        dfs(x,y-1)
        dfs(x+1,y)
        dfs(x,y+1)
        return True
    return False
answer=0
for i in range(N):
    for j in range(M):
        if dfs(i,j)==False:
            answer+=1
print(answer)
```
